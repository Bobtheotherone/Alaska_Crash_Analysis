```python
#make a script to clean the dataset

import pandas as pd

file_path1='Crash Level 13-17 (1).xlsx'
file_path2='Crash Level 09-12 (1).xlsx'

df1 = pd.read_excel(file_path1)
df2 = pd.read_excel(file_path2)

df1 = df1.drop(columns=[col for col in df1.columns if "Vehicle 2" in col or "Vehicle 3" in col])
df2 = df2.drop(columns=[col for col in df2.columns if "Vehicle 2" in col or "Vehicle 3" in col])

columns_in_13_17 = df1.columns.tolist()
columns_in_09_12 = df2.columns.tolist()

print("Columns in 09-12 that are not in 13-17:")
print(set(columns_in_09_12) - set(columns_in_13_17))
print()
print("Columns in 13-17 that are not in 09-12:")
print(set(columns_in_13_17) - set(columns_in_09_12))
print()
print("Common columns in both datasets:")
print(set(columns_in_13_17) & set(columns_in_09_12))
```
